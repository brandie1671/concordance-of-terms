```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.text import ConcordanceIndex

nltk.download('punkt')
```

```python
df = pd.read_csv('scientific_article.csv')
```

```python
# Handle NaN values in the 'title' column
df['title'] = df['title'].fillna('')

# Tokenize the titles
df['tokenized_title'] = df['title'].apply(lambda x: word_tokenize(str(x)))
```

```python
# Find concordance for the specific term 'appeal_to_reader'
term = 'appeal_to_reader'
concordance_idx = ConcordanceIndex(df['tokenized_title'].explode().reset_index()['tokenized_title'])
concordance_results = concordance_idx.offsets(term)

# Print the concordance lines
if concordance_results:
    for offset in concordance_results:
        start = max(0, offset - 5)
        end = min(len(df['tokenized_title'].explode()), offset + 5)
        context = list(df['tokenized_title'].explode().reset_index()['tokenized_title'][start:end])
        print(' '.join(context))
        print()
else:
    print(f"No occurrences of the term '{term}' found in titles.")
```
